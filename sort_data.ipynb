```python
import pandas as pd

metada_df = pd.read_csv('data/medical-diff-vqa-1.0.1/mimic_all.csv')
metada_df.shape
```

```python
vqa_df = pd.read_csv('data/medical-diff-vqa-1.0.1/mimic_pair_questions.csv')
vqa_df.shape
```

```python
metada_df_sorted = metada_df.sort_values(by=['subject_id', 'study_date'])
metada_df_sorted.to_csv('data/medical-diff-vqa-1.0.1/mimic_all_sorted.csv', index=False)
```

```python
vqa_df_sorted = vqa_df.sort_values(by=['subject_id', 'study_id'])
vqa_df_sorted.to_csv('data/medical-diff-vqa-1.0.1/mimic_pair_sorted.csv', index=False)
```
